```python
#input filenames
file1=input("Enter the source file:")
file2=input("Enter the destination file:")
f1=open(file1,"r")
f2=open(file2,"w")
i=0
for line in f1.readlines():
    i=i+1
    if i %2==0:
        continue
        f2.write(line)
f1.close()
f2.close()
print("odd lines copied sucessfully from",file1,"to",file2)
```

```python
import csv
fname=input("Enter the csv file name:")
with open(fname,"r")as f1:
    reader=csv.DictReader(f1)
    print("\nDepartment Details:")
    print("-----------")
    for row in reader:
        print("Deptld:",row['Deptld'],"\tDepName:",row['DepName'])
```

```python
import csv
data=[
    {"name":"anu","age":21,"marks":85},
    {"name":"saniya","age":22,"marks":90},
    {"name":"meera","age":20,"marks":80}
]
with open("dict.csv","w",newline="")as file:
          fieldnames=["name","age","marks"]
          writer=csv.DictWriter(file,fieldnames=fieldnames)
          writer.writeheader()
          writer.writerows(data)
print("Dictionary data written sucessfully!")

#step2
with open("dict.csv","r")as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
        
```
